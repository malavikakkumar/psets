### Question 1

Using the ``random`` library, create a random number of trials (e.g., between 10 - 20) of rts that randomly vary between 450 and 700 ms.

Write some code that:

* Prints the number of trials "completed".
* Calculates and prints the mean and standard deviation of the reaction times.
* Prints the fastest and slowest reaction times.

(Feel free to use the ``statisics`` library or use ``numpy`` if you prefer)

```python
import random, statistics

"""
I rewrote my original code as a function to generate a new RT list to use (repeatedly) in Question 4.
"""
def generate_rts(trials): # accepts as argument the number of trials required; user-entered
    rts = [random.randint(450,700) for trial in range(trials)] # adds a randomly generated rt within this range for each trial
    """
    above is equivalent to:
    trials = []
    for trial in range(trials):
    rts.append(random.randint(450,700))
    """
    print(rts) # functions as a check
    return rts

def rtStats(rts):
    """
    takes list of reaction times as input
    and returns calculated mean and standard deviation as output
    """
    mean_rt = statistics.mean(rts)
    stdev_rt = statistics.stdev(rts)
    return mean_rt, stdev_rt

def rtRank(rts):
    """
    takes list of reaction times as input
    checks list by element for max and min values
    """
    max_rt = 450 # predefined lower limit for rts in generated data, could also use 0
    min_rt = 700 # predefined upper limit for rts in generated data, could also use a ludicrously high number or just inf
    for rt in rts:
        if rt > max_rt: # if the value checked is greater than current max value
            max_rt = rt # assign checked value as max value
        if rt < min_rt: # if the value checked is lesser than current min value
            min_rt = rt # assign checked value as min value
    return min_rt, max_rt

def summarize_rts(rts):
    """
    this function is to automatically call rtStats and rtRank
    and print the returned values
    """
    print("Number of trials completed: ", len(rts))
    print("Mean and Standard Deviation of reaction times recorded: %f, %f" %(rtStats(rts)))
    print("Fastest and slowest reaction times recorded: %i, %i" %(rtRank(rts)))
    print("-----------------") # prints a separating line, so it looks nice :)

trials = random.randint(10, 20) # to randomly determine the number of trials, within the given range
rts = generate_rts(trials)
summarize_rts(rts)
```

### Question 2

Using the same list rts from Question 1, write code that:

* Creates a new list called trimmed_rts that includes only trials below 600 ms.
* Prints the number of trials removed.
* Prints the new mean reaction time.

```python
upper = 600 # a variable to define the upper limit

def rtTrim(rts, upper):
    trimmed_rts = [rt for rt in rts if rt < upper] # list comprehension method, similar to above, but with the addition of a conditional statement
    # can be explained as: FOR rt in rts, IF rt < upper THEN add rt to trimmed_rts
    trials_removed = len(rts) - len(trimmed_rts) # subtracts list lengths to determine number of variables removed
    trimmed_mean, trimmed_stdev = rtStats(trimmed_rts) # calls the rtStats function defined above but with new trimmed list, 
    # and stores returned values in two new variables
    print(trimmed_rts)
    print("Number of trials removed below %i ms: %i" %(upper, trials_removed))
    print("New mean of trimmed list: ", trimmed_mean)
    print("New standard deviation of trimmed list: ", trimmed_stdev)
    return trimmed_rts
    
trimmed_rts = rtTrim(rts, upper) # stores returned value (trimmed list) of rtTrim in a new list
```

### Question 4

Write some code that that:

* Generates multiple subjects RTs within one variable (can be stored in a list, dictionary, or numpy array)

* Uses your summarize_rts() function to print each participant’s mean reaction time.

* Stores each participant’s mean in a list.

* Prints the group mean across all participants.


```python
trials = random.randint(10, 20) # to randomly determine the number of trials, within the given range
participants = random.randint(10,25) # to randomly determine the number of participants within the given range
participant_rts = {} # empty dictionary to store rts list for each participant
mean_rts = [] # empty list to store mean rt of each participant

for ppt in range(1, participants + 1): 
# because the range() function starts from 0 by default, setting start as 1 
# because the range() function is not inclusive of the stop point, setting stop as (participants + 1) 
    participant_rts[ppt] = generate_rts(trials) # using the function defined in Question 1 to generate rts between 450-700 ms, given a certain number of trials
    # each generated list is stored in the dict participant_rts by the participant number
    print("For participant", ppt)
    summarize_rts(participant_rts[ppt])
    rt_mean = statistics.mean(participant_rts[ppt])
    mean_rts.append(rt_mean) # appends only calculated mean to mean_rts list

group_mean = statistics.mean(mean_rts)

print("Group mean of %i participants: %f" %(participants, group_mean))
```

### Question 5 

Create a class Participant that:

* Has attributes pid (ID) and rts (list of reaction times).

Includes methods:

* mean_rt() — returns the mean RT

* add_rt(rt) — adds a new RT

* num_trials() — returns how many trials they have

Then:

* Create an object for "S01" using the original data.

* Add a new reaction time of 505.

* Print the participant’s ID, number of trials, and updated mean.

```python
class Participant:

    def __init__(self, ID, rts_list):
        self.pid = ID
        self.rts = rts_list

    def mean_rt(self): # takes no arguments other than self since rts is already defined as part of the class
        return statistics.mean(self.rts)

    def add_rt(self, rt = 0): # takes as argument rt, which is a reaction time to be added to the existing rts_list
        self.rts.append(rt) # no need to return as rts is already defined as part of the class

    def num_trials(self): # takes no arguments other than self since rts is already defined as part of the class
        return len(self.rts) # the length of rts list = number of trials

participant_S01 = Participant('S01', rts)
participant_S01.add_rt(505)

print("Number of trials completed for participant %s: %i" %(participant_S01.pid, participant_S01.num_trials()))
print("Updated mean RT for participant %s: %f" %(participant_S01.pid, participant_S01.mean_rt()))
```

### Question 6

You want to quickly simulate one trial of your reaction-time task in PsychoPy.

Write a minimal PsychoPy script that:

* Displays a fixation cross (+) for 1 second.

* Then displays the word "GO!" until a key is pressed.

* Records and prints the reaction time in milliseconds.

(Hint: Use from psychopy import visual, core, event; a win, Clock, and event.waitKeys().)

```python
from psychopy import visual, core, event

# firstly, initialising a window and clock for the experiment
sampleWin = visual.Window(size = [800,600], color=[255,255,255], colorSpace='rgb255', fullscr=False, units='pix')
respClock = core.Clock()

fixation = visual.TextStim(sampleWin, text = '+', color=[0,0,0], colorSpace='rgb255', height=50) # fixation cross
fixation.draw() # draws TextStim fixation on sampleWin
sampleWin.flip()
core.wait(1.0) # tells the system to wait for 1 sec

prompt = visual.TextStim(sampleWin, text = 'GO!', color=[0,0,0], colorSpace='rgb255', height=100, bold=True) # creates textStim for text "GO!"
prompt.draw() # draws TextStim prompt on sampleWin
sampleWin.flip()
respClock.reset() # right after flipping, clock is reset to 0
event.waitKeys() # tells the system to wait until a key is pressed
rt = respClock.getTime() * 1000 # stops clock and records value as soon as the user responds with a keypress. This is the reaction time. 
# multiplying by 1000 to convert to milliseconds

rt_announcement = "Reaction time in ms:\n" + str(rt) # preparing to print RT to screen coherently
output = visual.TextStim(sampleWin, text = rt_announcement, color=[0,0,0], colorSpace='rgb255', height=50) # a new TextStim with RT text as defined above
output.draw() # draws TextStim output on sampleWin
sampleWin.flip()
core.wait(2.0) # tells the system to wait for 2 secs
```
