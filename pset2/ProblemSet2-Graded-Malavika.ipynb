_(b)_ Create a function that calculates the mean and standard deviation some list of values, and remove any values that exceed 2.5 SDs of the mean, recursively, until there are no outliers to remove. **Note:** The numbers to be trimmed should be an argument passed to the function. (/10)

```python
import statistics as stats

# defining a function that takes a list as an argument
# or at least, we're treating the argument as a list and hoping that's what the user actually passes lol
# I'm making a help documentation entry in the function to provide clarification for a potential user

def listStats(valuesList):
    """
    Calculates the mean and standard deviation of a list of numerical values,
    and recursively removes any values that exceed 2.5 SDs of the mean
    until there are no outliers to remove.
    
        Parameters
        ----------
        valuesList : LIST
            List of numerical values on which
            the calculations are to be done.
    
        Returns
        -------
        listMean : FLOAT
            FINAL MEAN CALCULATED FOR VALUESLIST.
        listSD : FLOAT
            FINAL STD DEV CALCULATED FOR VALUESLIST.
        outliers : INT
            A COUNT OF THE NUMBER OF REMOVED VALUES.
    """
    
    # initialising outlier count variable 
    outliers = 0

    # this loop will run as long as valuesList is True, i.e. not empty
    while valuesList:
        
        # these values will be re-calculated for each iteration of the loop
        listMean = stats.mean(valuesList)
        listSD = stats.stdev(valuesList)
        lower = listMean - (2.5 * listSD)
        upper = listMean + (2.5 * listSD)
        # creating a new list including all the values in the current version of valuesList
        # which do not exceed the lower and upper limit calculated for the current valuesList
        # this automatically excludes the outliers, of course
        valuesListNew = [value for value in valuesList if lower <= value <= upper]
        # calculating the number of excluded outliers in the current iteration by subtracting the lengths of both lists
        outliersNew = len(valuesList) - len(valuesListNew)
        # if there is no difference in lengths, this means there are no more outliers to be excluded
        # thus, a break command is used to end the loop prematurely
        if outliersNew == 0:
            break
        # if outliersNew is NOT 0, we add this count to the overall outliers
        outliers += outliersNew
        # and update valuesList to its newer version for the next iteration of the loop
        valuesList = valuesListNew

    # return the 3 values calculated within this function
    # valuesList will be updated independently as it was already defined external to the function
    return listMean, listSD, outliers
```

*(b)* Write a script that loops through each of the students, and then return's their letter grade based on this convention:

* A+: >=90
* 85 >= A <90
* 80 >= A- <85
* 76 >= B+ < 80

(/6)

```python
# creating an empty list to store all the calculated letter grade values
letterGrades = []

# as studentNames and studentGrades (the lists created in above step, holding values for student names and grades respectively)
# are both of the same length, and in the same order, the same index can be used to iterate through both lists simultaneously
# as the following loop iterates through each value of inx, we're able to move through both the lists
for inx in range(len(studentNames)):
    # getting the value of studentNames at the current index and assigning it to a dedicated variable
    studentName = studentNames[inx]
    # getting the value of studentGrades at the current index and assigning it to a dedicated variable
    studentGrade = studentGrades[inx]
    # the following if-elif-else checks which of the conditions the current value of studentGrade match
    # and assigns the corresponding value to a variable letterGrade
    if studentGrade >= 90:
        letterGrade = 'A+'
    elif 85 <= studentGrade < 90:
        letterGrade = 'A'
    elif 80 <= studentGrade < 85:
        letterGrade = 'A-'
    elif 76 <= studentGrade < 80:
        letterGrade = 'B+'
    # if the input is not between 76 and 90, a letter grade does not get assigned
    # the variable letterGrade will be of None type
    else:
        letterGrade = None
    # stores the current value of letterGrade to the letterGrades list
    # for future reference, in the same order as the studentNames and studentGrades list
    letterGrades.append(letterGrade)
    # prints the output from the current loop
    print(studentName, ":", studentGrade, letterGrade)
```

**Question 4:** Write a Python class called PersonalityProfile that stores a participant’s scores on each of the Big Five traits (Openness, Conscientiousness, Agreeableness, Extraversion, Neuroticism) on a scale from 1–5.

Your class should include:

An \_\_init\_\_ method that initializes the five traits, as well as the participant number.

A method _is_introvert()_ that returns True if the participant’s Extraversion score is less than 3, and False otherwise (feel free to make a similar method for all 5 traits).

**Bonus:** Create a method called _summary()_ that returns a string describing the participant’s strongest trait (the one with the highest score).


```python

class PersonalityProfile:

    # defining __init__ to initialise five traits + participant number, passed as arguments to the method by the user
    def __init__(self, ppt, openness, conscientious, agreeable, extravert, neurotic):
        # making a list of trait labels for future reference in this class
        self.traits = ["Openness", "Conscientiousness", "Agreeableness", "Extraversion", "Neuroticism"]
        # this conditional checks that all the values are between 1-5 and the participant number is non-negative
        if (1 <= openness <= 5 and 1 <= conscientious <= 5 and 1 <= agreeable <= 5 and 1 <= extravert <= 5 and 1 <= neurotic <= 5) and ppt > 0:
            # when true, the traits are initialised as class attributes and assigned the values passed by the user
            self.participantNo = ppt
            self.Openness = openness
            self.Conscientiousness = conscientious
            self.Agreeableness = agreeable
            self.Extraversion = extravert
            self.Neuroticism = neurotic
        else:
            # when false, prints an error message and the object (class instance) is of None type.
            self = None
            print("Please make sure trait values are between 1-5.")
            print("Please make sure participant number is higher than 0.")

    def is_introvert(self):
        # checks whether Extraversion is below 3, and if so indicates introversion
        if self.Extraversion < 3:
            # if so, the method is_introvert returns True
            return True
        else:
            return False

    def is_open(self):
        # checks whether Openness is equal to or above 3, and if so indicates openness is True
        if self.Openness >= 3:
            return True
        else:
            return False

    def is_conscientious(self):
        # checks whether Conscientiousness is equal to or above 3, and if so indicates conscientiousness is True
        if self.Conscientiousness >= 3:
            return True
        else:
            return False

    def is_agreeable(self):
        # checks whether Agreeableness is equal to or above 3, and if so indicates agreeableness is True
        if self.Agreeableness >= 3:
            return True
        else:
            return False

    def is_neurotic(self):
        # checks whether Neuroticism is equal to or above 3, and if so indicates neuroticism is True
        if self.Neuroticism >= 3:
            return True
        else:
            return False

    def summary(self):
        # the goal is to return the highest scored trait
        maxValue = 0 
        # maxValue is the variable that will store the highest score
        # currently holds the smallest possible value (that will not be contained in the trait scores)
        print()
        for trait in self.traits: # self.traits is a list of all the trait labels in string form
            # this loop iterates through each trait in self.traits
            # trait is used as the key to access only the trait class attributes (and exclude participant no., etc.)
            currentValue = self.__dict__[trait] # stores the score for the current trait
            print(trait + ":", currentValue) # and prints it
            if currentValue >= maxValue: # if the score for the current trait is higher than the last stored maxValue
                maxValue = currentValue # the maxValue is assigned the score stored in currentValue
        # since there may be more than one trait with the same score
        # the following appends all traits with the same score to a common list called maxTraits
        maxTraits = [trait for trait in self.traits if self.__dict__[trait] == maxValue]
        print("Participant's strongest trait is:")
        return maxTraits

# initialising an instance of the PersonalityProfile class
Prufrock = PersonalityProfile(1, 3, 5, 1, 2, 5)
# calling methods is_introvert() and summary() on Prufrock
print(Prufrock.is_introvert())
Prufrock.summary()
```
